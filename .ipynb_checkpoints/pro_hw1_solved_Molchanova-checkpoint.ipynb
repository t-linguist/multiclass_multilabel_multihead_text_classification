```python

# ---- Ваш код здесь ----
print("""
    Загружаем выбранную LLM
    В некоторых случаях может понадобиться !pip install -U accelerate bitsandbytes transformers
""")

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, GenerationConfig

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model_name = "IlyaGusev/saiga_yandexgpt_8b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map="auto"
)
generation_config = GenerationConfig.from_pretrained(model_name)
model.generation_config = generation_config
print(generation_config)
# ---- Конец кода ----
```

```python
# Промпт


# ---- Ваш код здесь ----
prompt_template = (
    """Ты — редактор новостей. Тебе дан следующий текст новостей:
{text}
Определи по шкале от 0 до 1 степень принадлежности текста новости к следующим тематикам:
- политика;
- экономика;
- культура.
Ответ оформи в виде json-файла: {{"politics": float rounded to two decimal places, "economy": float rounded to two decimal places, "culture": float rounded to two decimal places}}
"""
    )
# ---- Конец кода ----
```

```python
 # Функция разметки

# @torch.no_grad() # можно включить декоратор
def annotate(text: str, max_length:int=512) -> dict:
    """
    Формат markup следующий
        markup = {
        'politics': 0/1,
        'economy': 0/1,
        'culture': 0/1,
        'corrupted': 0/1, # сломалось ли что-то при получении разметки
        'generation': llm_generation # чтобы можно было раздебажить закоррапченные разметки
    }
    """
    try:
        # вставляем текст в шаблон, обрезаем на всякий чтоб не падало по памяти лишний раз
        prompt_input = prompt_template.format(text=text[:1500]) # обрезаем слишком длинные новости


        # может понадобиться, может нет -- зависит от выбранной вами модели
        # prompt = tokenizer.apply_chat_template([{
        #     "role": "user",
        #     "content": prompt_input
        # }], tokenize=False, add_generation_prompt=True)

        # markup = {}
        # with torch.no_grad():
        #     # inference tokenizer, model ...

        #     del ... # удаляем ненужные переменные
        #     torch.cuda.empty_cache()

        #     return markup

        # ---- Ваш код здесь ----
        print("""
            непосредственно разметка одного текста text с помощью выбранной LLM
        """)
        prompt = tokenizer.apply_chat_template([{
            "role": "user",
            "content": prompt_input
        }], tokenize=False, add_generation_prompt=True)
        data = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)
        data = {k: v.to(model.device) for k, v in data.items()}
        data.pop("token_type_ids", None)
        output_ids = model.generate(**data, max_new_tokens=96, generation_config=generation_config)[0]
        output_ids = output_ids[len(data["input_ids"][0]):]
        markup = tokenizer.decode(output_ids, skip_special_tokens=True).strip()
        print(markup)
        # ---- Конец кода ----
        return markup

    except torch.cuda.OutOfMemoryError:
        print("CUDA OOM. Освобождаем память...")
        torch.cuda.empty_cache()
        return {
            'politics': 0, 'economy': 0, 'culture': 0,
            'corrupted': 1,
            'generation': 'OOM_ERROR'
        }

first_text = df["title"][0] + "\n" + df["text"][0]
print(annotate(first_text))
```

```python
from tqdm import tqdm

gen_results = []
for idx in tqdm(range(1000)):
    try:
        gen_results.append(annotate(df["title"][idx] + "\n" + df["text"][idx]))
    except Exception:
        gen_results.append({'politics': 0, 'economy': 0, 'culture': 0})
gen_results
```

```python
import re
import json

def extract_json(line: str):
    regexp = re.findall(r"{.*?}", line, flags=re.S)[0]
    return json.loads(regexp)

extract_json(gen_results[0])
```

```python
gen_results_clean = [extract_json(res) for res in gen_results]
gen_results_clean
```

```python
# ---- Ваш код здесь ----
print("""
    прокачиваем в цикле выбранную LLM для разметки данных через функцию annotate, добавляем разметку в исходный датасет и сохраняем в файл
""")
marked_df = df.iloc[:1000].copy(deep=True)
marked_df["politics"] = [res["politics"] for res in gen_results_clean]
marked_df["economy"] = [res["economy"] for res in gen_results_clean]
marked_df["culture"] = [res["culture"] for res in gen_results_clean]
marked_df["generation"] = gen_results
marked_df.sample(7)
# ---- Конец кода ----

```

```python
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit


def prepare_data(df):
    """
    Подготовка данных для multiclass multilabel классификации
    """
    X = df[["title", "text"]].values
    target_columns = ["politics", "economy", "culture"]
    y = df[target_columns].values
    
    # Проверяем распределение исходных данных
    print("Original distribution:")
    for i, col in enumerate(target_columns):
        print(f"{col}: min={y[:, i].min():.3f}, max={y[:, i].max():.3f}, "
              f"mean={y[:, i].mean():.3f}")
    
    # Создаем бинарные метки для стратификации
    y_binary = (y > 0.5).astype(int)
    
    # Многоуровневое стратифицированное разбиение
    msss = MultilabelStratifiedShuffleSplit(
        n_splits=1, 
        test_size=0.2, 
        random_state=42
    )
    
    for train_idx, test_idx in msss.split(X, y_binary):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
    
    return X_train, X_test, y_train, y_test
```

```python
# ---- Ваш код здесь ----
print("EDA и разделение на трейн-тест")

X_train, X_test, y_train, y_test = prepare_data(df)
# ---- Конец кода ----
```

### Классы для обучения и подготовки данных

```python
import torch
import torch.nn as nn
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
import numpy as np
from tqdm import tqdm

import torch
import torch.nn as nn
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import numpy as np
from tqdm import tqdm
import os
import joblib


# ---- Ваш код здесь ----
print("Пишем классы, обучаем")

X_train = np.loadtxt('x_train.csv', dtype=str, delimiter="\t", encoding="utf-8")
X_test = np.loadtxt('x_test.csv', dtype=str, delimiter="\t", encoding="utf-8")
y_train = np.loadtxt('y_train.csv', delimiter=" ", encoding="utf-8")
y_test = np.loadtxt('y_test.csv', delimiter=" ", encoding="utf-8")


class NewsDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=256):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length', 
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(), 
            'labels': torch.FloatTensor(self.labels[idx])
        }


class MultiHeadNewsClassifier(nn.Module):
    def __init__(self, base_model_name="cointegrated/rubert-tiny2", initial_categories=None):
        super().__init__()
        self.bert = AutoModel.from_pretrained(base_model_name)
        self.hidden_size = self.bert.config.hidden_size
        
        self.category_heads = nn.ModuleDict()
        
        if initial_categories:
            for category in initial_categories:
                self.add_category_head(category)
    
    def add_category_head(self, category_name):
        head = nn.Sequential(
            nn.Dropout(0.1),
            nn.Linear(self.hidden_size, 64),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(64, 1)
        )
        self.category_heads[category_name] = head
        print(f"Добавлена голова для: {category_name}")
    
    def forward(self, input_ids, attention_mask, categories=None):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.last_hidden_state[:, 0, :]
        
        if categories is None:
            categories = list(self.category_heads.keys())
        
        logits = {}
        for category in categories:
            logits[category] = self.category_heads[category](pooled_output)
        
        return logits


class MultiHeadTrainingPipeline:
    def __init__(self, model_name="cointegrated/rubert-tiny2"):
        self.model_name = model_name
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        print(f"Используется устройство: {self.device}")
        
    def prepare_data(self, X_train, X_test, y_train, y_test, initial_categories):
        """ПОДГОТОВКА ДАННЫХ И СОЗДАНИЕ DATALOADER'ОВ"""
        print("Подготовка данных и создание DataLoader'ов...")
        
        # СОЗДАЕМ DATASET'Ы
        self.train_dataset = NewsDataset(X_train, y_train, self.tokenizer)
        self.val_dataset = NewsDataset(X_test, y_test, self.tokenizer)
        
        # СОЗДАЕМ DATALOADER'Ы
        self.train_loader = DataLoader(self.train_dataset, batch_size=8, shuffle=True)
        self.val_loader = DataLoader(self.val_dataset, batch_size=8, shuffle=False)
        
        print(f"Train samples: {len(self.train_dataset)}")
        print(f"Val samples: {len(self.val_dataset)}")
        print(f"Train batches: {len(self.train_loader)}")
        print(f"Val batches: {len(self.val_loader)}")
        
        return self.train_loader, self.val_loader
    
    def initialize_model(self, initial_categories):
        """Инициализация модели и оптимизаторов"""
        self.model = MultiHeadNewsClassifier(
            base_model_name=self.model_name,
            initial_categories=initial_categories
        ).to(self.device)
        
        # Создаем оптимизаторы для каждой головы
        self.category_optimizers = {}
        for category in initial_categories:
            self.category_optimizers[category] = torch.optim.AdamW(
                self.model.category_heads[category].parameters(), 
                lr=1e-3
            )
        
        # Оптимизатор для backbone
        self.backbone_optimizer = torch.optim.AdamW(
            self.model.bert.parameters(), 
            lr=2e-5
        )
        
        self.criterion = nn.BCEWithLogitsLoss()
        
        print(f"Модель инициализирована с категориями: {initial_categories}")
    
    def train_category(self, category, epoch_num, freeze_backbone=False):
        """Обучение одной категории"""
        self.model.train()
        
        if freeze_backbone:
            self.model.bert.eval()
        
        total_loss = 0
        progress_bar = tqdm(self.train_loader, desc=f'Epoch {epoch_num} [{category}]')
        
        for batch in progress_bar:
            # Zero gradients
            self.category_optimizers[category].zero_grad()
            if not freeze_backbone:
                self.backbone_optimizer.zero_grad()
            
            input_ids = batch['input_ids'].to(self.device)
            attention_mask = batch['attention_mask'].to(self.device)
            
            # Получаем индекс категории для labels
            category_idx = list(self.model.category_heads.keys()).index(category)
            labels = batch['labels'][:, category_idx].to(self.device)
            
            # Forward pass
            logits_dict = self.model(input_ids, attention_mask, categories=[category])
            logits = logits_dict[category].squeeze()
            
            loss = self.criterion(logits, labels)
            loss.backward()
            
            self.category_optimizers[category].step()
            if not freeze_backbone:
                self.backbone_optimizer.step()
            
            total_loss += loss.item()
            progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        return total_loss / len(self.train_loader)
    
    def train_all_categories(self, epoch_num, freeze_backbone=False):
        """Обучение всех категорий вместе"""
        self.model.train()
        
        if freeze_backbone:
            self.model.bert.eval()
        
        total_loss = 0
        progress_bar = tqdm(self.train_loader, desc=f'Epoch {epoch_num} [All]')
        
        for batch in progress_bar:
            # Zero gradients
            for optimizer in self.category_optimizers.values():
                optimizer.zero_grad()
            if not freeze_backbone:
                self.backbone_optimizer.zero_grad()
            
            input_ids = batch['input_ids'].to(self.device)
            attention_mask = batch['attention_mask'].to(self.device)
            labels = batch['labels'].to(self.device)
            
            # Forward pass для всех категорий
            logits_dict = self.model(input_ids, attention_mask)
            
            total_batch_loss = 0
            for i, category in enumerate(logits_dict.keys()):
                category_logits = logits_dict[category].squeeze()
                category_labels = labels[:, i]
                loss = self.criterion(category_logits, category_labels)
                total_batch_loss += loss
            
            total_batch_loss.backward()
            
            for optimizer in self.category_optimizers.values():
                optimizer.step()
            if not freeze_backbone:
                self.backbone_optimizer.step()
            
            total_loss += total_batch_loss.item()
            progress_bar.set_postfix({'loss': f'{total_batch_loss.item():.4f}'})
        
        return total_loss / len(self.train_loader)
    
    def validate(self):
        """Валидация модели"""
        self.model.eval()
        total_loss = 0
        all_predictions = []
        all_labels = []
        
        with torch.no_grad():
            for batch in tqdm(self.val_loader, desc='Validation'):
                input_ids = batch['input_ids'].to(self.device)
                attention_mask = batch['attention_mask'].to(self.device)
                labels = batch['labels'].to(self.device)
                
                logits_dict = self.model(input_ids, attention_mask)
                
                batch_loss = 0
                batch_predictions = []
                
                for i, category in enumerate(logits_dict.keys()):
                    category_logits = logits_dict[category].squeeze()
                    category_labels = labels[:, i]
                    loss = self.criterion(category_logits, category_labels)
                    batch_loss += loss.item()
                    
                    probabilities = torch.sigmoid(category_logits)
                    batch_predictions.append(probabilities.cpu().numpy())

                total_loss += batch_loss
                all_predictions.append(np.column_stack(batch_predictions))
                all_labels.append(labels.cpu().numpy())
        
        avg_loss = total_loss / len(self.val_loader)
        all_predictions = np.vstack(all_predictions)
        all_labels = np.vstack(all_labels)
        
        return avg_loss, all_predictions, all_labels
    
    def calculate_metrics(self, predictions, labels, threshold=0.5):
        """Расчет метрик"""
        from sklearn.metrics import f1_score, precision_score, recall_score
        
        binary_preds = (predictions > threshold).astype(int)
        
        metrics = {
            'f1_micro': f1_score(labels, binary_preds, average='micro'),
            'f1_macro': f1_score(labels, binary_preds, average='macro'),
            'precision_micro': precision_score(labels, binary_preds, average='micro'),
            'recall_micro': recall_score(labels, binary_preds, average='micro'),
        }
        
        # Метрики по категориям
        for i, category in enumerate(self.model.category_heads.keys()):
            metrics[f'f1_{category}'] = f1_score(labels[:, i], binary_preds[:, i])
        
        return metrics
# ---- Конец кода ----
```

```python
# Инициализация пайплайна
pipeline = MultiHeadTrainingPipeline()

# ПОДГОТОВКА ДАННЫХ И СОЗДАНИЕ DATALOADER'ОВ
initial_categories = ['politics', 'economy', 'culture']
train_loader, val_loader = pipeline.prepare_data(X_train, X_test, y_train, y_test, initial_categories)

# Инициализация модели
pipeline.initialize_model(initial_categories)

# История обучения
history = {
    'train_loss': [],
    'val_loss': [],
    'val_metrics': []
}

# Цикл обучения
epochs = 3
for epoch in range(epochs):
    print(f"Эпоха {epoch + 1}/{epochs}")    
    # Обучение всех категорий
    train_loss = pipeline.train_all_categories(epoch + 1)
    
    # Валидация
    val_loss, val_predictions, val_labels = pipeline.validate()
    val_metrics = pipeline.calculate_metrics(val_predictions, val_labels)
    
    # Сохранение истории
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['val_metrics'].append(val_metrics)
    
    print(f"\nЭпоха {epoch + 1} завершена:")
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Val Loss: {val_loss:.4f}")
    print("Метрики валидации:")
    for metric, value in val_metrics.items():
        print(f"  {metric}: {value:.4f}")
print("\nОбучение завершено!")
```
